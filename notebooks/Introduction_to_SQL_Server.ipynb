####**2. Commandez ! Commande!**

+ ***Les tableaux comprennent des lignes et des colonnes, et vous verrez généralement les résultats de votre requête affichés sous forme de tableau. ####**3. Trier par - ordre croissant**

+ ***Pour ordonner nos résultats de requête, nous écrivons `ORDER BY` sous `FROM`, puis listons les colonnes que nous voulons trier. Ici, nous trions par `year_intro`, puis la colonne `product_id`. ####**4. Trier par - ordre décroissant**

+ ***Ici, nous ajoutons `DESC` après la colonne year_intro dans la clause `ORDER BY`. Cela trie les résultats par ordre décroissant.***

####**5. Trier par - plusieurs colonnes, différentes directions**

+ ***Nous pouvons trier sur plusieurs colonnes. Sur la gauche, les colonnes sont d'abord triées par `year_intro` dans l'ordre décroissant, puis par canaux dans l'ordre croissant. `Year_intro` est répertorié en premier dans la clause `ORDER BY`, il est donc prioritaire lors du tri des lignes. Ensuite, la colonne des chaînes est triée par ordre croissant, pour chaque année. À droite, les deux colonnes sont triées par ordre décroissant.***

+ ***Dans un exercice précédent, vous avez écrit une WHERErequête distincte pour déterminer la quantité de demande perdue pour une région spécifique. Nous ne voudrions pas avoir à écrire des requêtes individuelles pour chaque région. Heureusement, vous n'avez pas besoin d'écrire des requêtes individuelles pour chaque région. Avec GROUP BY, vous pouvez obtenir une somme de toutes les valeurs uniques pour la colonne que vous avez choisie, en une seule fois.***

+ *Vous reviendrez au tableau grid ici et calculerez la demande perdue totale pour toutes les régions.*

####**Instructions**

+ Sélectionnez nerc_regionet la somme de demand_loss_mwpour chaque région.
+ Exclure les valeurs où demand_loss_mwest NULL.
+ Regroupez les résultats par nerc_region.
+ Classer par ordre décroissant de demand_loss.

```python
-- Select the region column
SELECT
  nerc_region,
  -- Sum the demand_loss_mw column
  SUM(demand_loss_mw) AS demand_loss
FROM
  grid
  -- Exclude NULL values of demand_loss
WHERE
  demand_loss_mw IS NOT NULL
  -- Group the results by nerc_region
GROUP BY
  nerc_region
  -- Order the results in descending order of demand_loss
ORDER BY
  demand_loss DESC;
```

####**Ayant**
+ ***WHERE est utilisé pour filtrer les lignes avant tout regroupement. Une fois que vous avez effectué une opération de regroupement, vous souhaiterez peut-être restreindre davantage le nombre de lignes renvoyées. C'est un travail pour HAVING. Dans cet exercice, vous allez modifier une requête existante pour utiliser HAVING, afin que seuls les résultats dont la somme est supérieure 10000 soient renvoyés.***

####**Instructions**

+ Modifiez la requête fournie pour supprimer la clause WHERE.
+ Remplacez-la par une clause HAVING afin que seuls les résultats dont le total demand_loss_mw est supérieur à 10000 soient renvoyés.

```python
SELECT
  nerc_region,
  SUM (demand_loss_mw) AS demand_loss
FROM
  grid
  -- Remove the WHERE clause
WHERE demand_loss_mw  IS NOT NULL
GROUP BY
  nerc_region
  -- Enter a new HAVING clause so that the sum of demand_loss_mw is greater than 10000
HAVING
  SUM(demand_loss_mw) > 10000
ORDER BY
  demand_loss DESC;
```

***Bien joué. Essayez de vous rappeler que `WHERE est appliqué avant GROUP BY` et filtre sur les lignes. `HAVING est appliqué après GROUP BY et filtre les groupes ou agrégats que vous avez créés dans la requête`.***

####**2. Deux requêtes**

+ ***Voici une requête qui renvoie un petit ensemble de résultats à partir de la table album. Alors que l'exemple interroge la même table, dans la vraie vie, nous pouvons avoir des données dans des tables séparées, qui partagent la même structure, que nous devons combiner en un seul ensemble de résultats. Qu'entendons-nous par la même structure ? Cela signifie que nous nous attendons à ce que les deux tables aient le même nombre de colonnes, les colonnes sont répertoriées dans le même ordre et ont des types de données similaires.***

####**3. Combiner les résultats**

+ ***La réponse est d'utiliser UNION. Nous écrivons la première requête, puis UNION sur une nouvelle ligne, puis nous écrivons la deuxième requête en dessous. ####**6. Résumé**

####**1. Insérer, mettre à jour, supprimer**

####**2. INSÉRER**

+ ***Une fois que nous avons créé un tableau, nous devrons y saisir des données. Une façon de procéder consiste à utiliser une instruction `INSERT`. Cela commence par les mots clés `INSERT INTO`, suivis du nom de la table. Nous tapons ensuite les noms de colonne dans lesquels nous voulons insérer des données, entre parenthèses. Nous devons séparer chaque nom de colonne par une virgule. Ensuite, nous tapons le mot clé `VALUES`, avant d'ouvrir à nouveau les parenthèses, et de taper les valeurs individuelles que vous souhaitez saisir, à nouveau, chacune séparée par une virgule. Ensuite, nous fermons la parenthèse pour terminer l'instruction `INSERT`.***

####**3. INSÉRER SÉLECTIONNER**

+ ***Une autre méthode que nous pourrions utiliser pour remplir une table consiste à effectuer un `INSERT SELECT`. Cela commence comme un `INSERT` - nous utilisons `INSERT INTO` suivi du nom de la table et des noms de colonne. Au lieu de taper `VALUES`, nous tapons `SELECT`, puis les noms des colonnes que nous voulons sélectionner, et le nom de la table source. Nous pouvons également appliquer n'importe quelle condition `WHERE` à ce stade. C'est toujours une bonne idée d'être précis sur les noms des colonnes dans lesquelles nous voulons insérer, et les noms correspondants des colonnes à partir desquelles nous voulons `SELECT`. De plus, nous devons veiller à ce que les noms des colonnes source et destination soient dans le bon ordre.***

####**4. MISE À JOUR**

+ ***Nous pouvons modifier les enregistrements à l'aide d'une instruction `UPDATE`. Cela commence par le mot-clé `UPDATE`, suivi du nom de la table. Ceci est suivi du mot-clé SET, de la colonne que nous voulons mettre à jour, de la valeur que nous voulons changer pour l'enregistrement et, surtout, d'une clause WHERE avec les conditions qui identifient les lignes spécifiques que nous voulons mettre à jour. N'oubliez pas la clause `WHERE`, sinon nous mettrons à jour TOUTES les valeurs de la colonne. Si nous voulons mettre à jour plus d'une colonne à la fois, nous listons simplement les colonnes supplémentaires et leurs nouvelles valeurs respectives***

####**5. SUPPRIMER**

+ ***Il est très facile de SUPPRIMER dans SQL Server. Sachez cependant que nous ne serons pas invités à confirmer - il n'y a pas de "êtes-vous sûr?" invites, donc lorsque nous exécutons une instruction de suppression, cela aura lieu immédiatement. La syntaxe est simplement DELETE FROM, le nom de la table et la clause WHERE pour spécifier les enregistrements à supprimer. Assurez-vous de tester d'abord la clause WHERE. Une autre méthode consiste à utiliser l'instruction `TRUNCATE TABLE`. Cela n'accepte pas ou n'exige pas une clause WHERE - cela supprimera toutes les données de toutes les colonnes à la fois.***

Vous pouvez le faire en créant une table temporaire qui reste dans la base de données jusqu'au redémarrage de SQL Server. Dans ce dernier exercice, vous sélectionnerez la piste la plus longue de chaque album et l'ajouterez dans une table temporaire que vous créerez dans le cadre de la requête.***

####**Instructions**

+ Insérez des données via une instruction SELECT dans une table temporaire appelée #maxtracks.
+ Joindre album à artist using artist_id, et track à album using album_id.
+ Exécutez l'instruction finale SELECT pour récupérer toutes les colonnes de votre nouvelle table.

```python
SELECT  album.title AS album_title,
  artist.name as artist,
  MAX(track.milliseconds / (1000 * 60) % 60 ) AS max_track_length_mins
-- Name the temp table #maxtracks
INTO #maxtracks
FROM album
-- Join album to artist using artist_id
INNER JOIN artist ON album.artist_id = artist.artist_id
-- Join track to album using album_id
INNER JOIN track  ON track.album_id = album.album_id
GROUP BY artist.artist_id, album.title, artist.name,album.album_id
-- Run the final SELECT query to retrieve the results from the temporary table
SELECT album_title, artist, max_track_length_mins
FROM  #maxtracks
ORDER BY max_track_length_mins DESC, artist;
```
