```python
DECLARE
	@SomeTime DATETIME2(7) = SYSUTCDATETIME();

-- Retrieve the year, month, and day
SELECT
	YEAR(@SomeTime) AS TheYear,
	MONTH(@SomeTime) AS TheMonth,
	DAY(@SomeTime) AS TheDay;
```

```python
DECLARE
	@LeapDay DATETIME2(7) = '2012-02-29 18:00:00';

-- Fill in the date parts and intervals as needed
SELECT
	DATEADD(DAY, -1, @LeapDay) AS PriorDay,
	DATEADD(DAY, 1, @LeapDay) AS NextDay,
    -- For leap years, we need to move 4 years, not just 1
	DATEADD(YEAR, -4, @LeapDay) AS PriorLeapYear,
	DATEADD(YEAR, 4, @LeapDay) AS NextLeapYear,
	DATEADD(YEAR, -1, @LeapDay) AS PriorYear;
```

```python
DECLARE
	@PostLeapDay DATETIME2(7) = '2012-03-01 18:00:00';

-- Fill in the date parts and intervals as needed
SELECT
	DATEADD(DAY, -1, @PostLeapDay) AS PriorDay,
	DATEADD(DAY, 1, @PostLeapDay) AS NextDay,
	DATEADD(YEAR, -4, @PostLeapDay) AS PriorLeapYear,
	DATEADD(YEAR, 4, @PostLeapDay) AS NextLeapYear,
	DATEADD(YEAR, -1, @PostLeapDay) AS PriorYear,
    -- Move 4 years forward and one day back
	DATEADD(DAY, -1, DATEADD(YEAR, 4, @PostLeapDay)) AS NextLeapDay,
    DATEADD(DAY, -2, @PostLeapDay) AS TwoDaysAgo;
```

```python
DECLARE
	@PostLeapDay DATETIME2(7) = '2012-03-01 18:00:00',
    @TwoDaysAgo DATETIME2(7);

SELECT
	@TwoDaysAgo = DATEADD(DAY, -2, @PostLeapDay);

SELECT
	@TwoDaysAgo AS TwoDaysAgo,
	@PostLeapDay AS SomeTime,
    -- Fill in the appropriate function and date types
	DATEDIFF(DAY, @TwoDaysAgo, @PostLeapDay) AS DaysDifference,
	DATEDIFF(HOUR, @TwoDaysAgo, @PostLeapDay) AS HoursDifference,
	DATEDIFF(MINUTE, @TwoDaysAgo, @PostLeapDay) AS MinutesDifference;
```

####**Arrondir les dates**

+ ***SQL Server ne dispose pas d'un moyen intuitif d'arrondir au mois, à l'heure ou à la minute. Vous pouvez cependant combiner les fonctions `DATEADD() et DATEDIFF()`pour effectuer cet arrondi.***

+ ***Pour arrondir la date 1914-08-16 à l'année, nous appellerions `DATEADD(YEAR, DATEDIFF(YEAR, 0, '1914-08-16'), 0)`.***

+ ***Pour arrondir la date au mois inférieur, nous appellerions `DATEADD(MONTH, DATEDIFF(MONTH, 0, '1914-08-16'), 0)`.***

+ *Cela fonctionne également pour plusieurs autres parties de date.*

####**Instructions**

+ Utilisez `DATEADD() et DATEDIFF()` en conjonction avec les parties de date pour arrondir notre heure au jour, à l'heure et à la minute.

```python
DECLARE
	@SomeTime DATETIME2(7) = '2018-06-14 16:29:36.2248991';

-- Fill in the appropriate functions and date parts
SELECT
	DATEADD(DAY, DATEDIFF(DAY, 0, @SomeTime), 0) AS RoundedToDay,
	DATEADD(HOUR, DATEDIFF(HOUR, 0, @SomeTime), 0) AS RoundedToHour,
	DATEADD(MINUTE, DATEDIFF(MINUTE, 0, @SomeTime), 0) AS RoundedToMinute;
```

***Super! Une chose à noter est que `DATEDIFF()` renvoie un type entier, il peut donc déborder si vous essayez d'arrondir en secondes. Il pourrait déborder en arrondissant aux minutes, mais ce sera après l'an 5000, donc nous sommes en sécurité pour l'instant.***

####**2. Fonctions de formatage**

+ ***SQL Server nous propose trois options de formatage : `CAST(), CONVERT() et FORMAT()`.***

####**3. La fonction CAST()**

+ ***La fonction `CAST()` est présente dans SQL Server depuis plus de deux décennies. Il permet de convertir un type de données en un autre type de données, tel qu'un entier en décimal. Lors de l'utilisation de `CAST()`, nous n'avons aucun contrôle sur la façon dont la fonction transforme nos dates en chaînes. Du côté positif, cependant,` CAST ()` fait partie de la norme ANSI SQL, ce qui signifie que presque toutes les bases de données qui implémentent SQL l'auront.***

####**4. Utilisation de la fonction CAST()**

+ ***Prenons un exemple de casting. Pour chacune des trois variables affichées, nous allons effectuer une opération `CAST()`. Nous allons transformer SomeDate en chaîne, SomeString en DATETIME2 et OldDateTime en chaîne. Comme vous pouvez le voir, la fonction `CAST()` convertit avec succès chacun d'entre eux, renvoyant respectivement une chaîne, une date et une chaîne "bien formatée". La façon dont il est formaté dépend de vos préférences personnelles !***

####**5. La fonction CONVERT()**

+ ***La fonction `CONVERT()` remonte également au moins à SQL Server 2000. Elle est utile pour la conversion entre les types de données, tout comme `CAST()`. Contrairement à `CAST()`, cependant, nous obtenons des informations sur l'apparence de nos dates lorsque nous les convertissons en chaînes. La fonction `CONVERT()` est spécifique à T-SQL, ce qui signifie que nous ne devons pas nous attendre à la trouver en dehors de SQL Server.***

####**6. Utilisation de la fonction CONVERT()**

+ ***Voyons comment fonctionne la fonction `CONVERT()` en utilisant la date déclarée indiquée. La fonction `CONVERT()` prend trois paramètres d'entrée : un type de données, une entrée et un style facultatif. Voyons comment certains de ces styles modifient nos cordes. Le style 0 imprime le texte de la même manière que `CAST()` le fait pour un type DATETIME. En utilisant les styles 1 ou 101, nous pouvons imprimer les dates au format commun des États-Unis. La différence est qu'avec 1, nous obtenons une année à deux chiffres et 101 imprime une année à quatre chiffres. Un autre exemple consiste à utiliser le style 120, qui imprime dans un format DATETIME2 standard à la seconde, appelé standard canonique ODBC, en ignorant les millisecondes et les microsecondes.***

####**7. Exemples de styles CONVERT()**

+ ***SQL Server prend en charge quelques dizaines de styles de conversion distincts. On a déjà vu le format américain, mais ce n'est pas le seul format national. Nous pouvons utiliser 3 ou 103 pour imprimer au format national britannique/français. Les autres formats affichés incluent l'allemand, le japonais et l'italien. Pour l'interopérabilité, les normes ISO ou ODBC sont recommandées. Nous recommandons la norme ISO8601 pour une interopérabilité optimale entre les systèmes et les plates-formes.***

####**8. La fonction FORMAT()**

+ ***La fonction `FORMAT()` est arrivée dans SQL Server 2012. Cela ajoute beaucoup plus de flexibilité dans les rapports que `CAST() ou CONVERT()`, mais est également spécifique à SQL Server. Il utilise le framework dot-NET pour la conversion et est monothread, il peut donc être plus lent que `CAST() ou CONVERT()`.***

####**9. Utilisation de la fonction FORMAT()**

+ ***`FORMAT()` prend trois paramètres : une entrée, un code de format et une culture facultative. En utilisant différentes cultures, nous pouvons imprimer des dates pour différents paramètres régionaux. Nous pouvons également spécifier un format personnalisé !***

####**10. Quelle est la lenteur de FORMAT() ?**

+ ***La fonction `FORMAT()` peut être plus lente que `CAST() ou CONVERT()`. Dans l'exemple illustré, nous pouvons voir qu'à partir d'environ 100 000 lignes, `FORMAT()` devient plus lent que l'une ou l'autre fonction. Dans d'autres exemples, `FORMAT()` devient pratiquement inutilisable dans la plage de 50 000 ou 100 000 enregistrements. Pour les ensembles de données de moins de quelques milliers de lignes, un délai d'utilisation de `FORMAT()` n'est probablement pas perceptible, mais les performances sont un élément à garder à l'esprit lorsque vous pensez à l'évolutivité.***

####**2. Dates des pièces**

+ ***SQL Server dispose de six fonctions pour créer des dates et des heures à partir de composants : la série « à partir de composants ». Le premier est `DATEFROMPARTS()`, qui prend des valeurs entières pour l'année, le mois et le jour et renvoie un type `DATE`. Le type de données DATETIME classique a sa propre fonction, `DATETIMEFROMPARTS()` qui descend jusqu'à une granularité de 3 millisecondes. Pour plus de précision, nous pouvons construire un `DATETIME2` à partir de parties en utilisant la fonction `DATETIME2FROMPARTS()`. Notez que ses entrées sont la combinaison définie de `DATEFROMPARTS() et DATETIMEFROMPARTS()`. Il existe une fonction héritée, `SMALLDATETIMEFROMPARTS()`. À ce point, nous vous recommandons d'éviter le type de données `SMALLDATETIME` et d'utiliser simplement `DATETIME2` à la place. Enfin, `DATETIMEOFFSETFROMPARTS()` prend les champs `DATETIME2` et ajoute des entrées pour les décalages d'heure et de minute par rapport à l'heure UTC afin que nous puissions spécifier le fuseau horaire ainsi que l'heure.***

####**1. Travailler avec des décalages**

+ ***L'une des parties les plus délicates de la gestion des dates concerne les décalages. Les types de données de date et d'heure classiques comme `DATETIME et DATETIME2` n'ont pas de concept intégré de fuseau horaire, donc une date qui lit 15 heures peut être en UTC ou dans un autre fuseau horaire.***

####**2. Anatomie d'un DATETIMEOFFSET**

+ ***Dans les cas où connaître le fuseau horaire est important, SQL Server a un autre type de date : le `DATETIMEOFFSET`. Le type `DATETIMEOFFSET` est composé de trois composants : `une date, une heure et un décalage UTC`. Tout cela prend 10 octets d'espace contre 8 octets pour un DATETIME ou 6-8 octets (selon le niveau de précision temporelle) pour un `DATETIME2`.***

####**3. Anatomie d'un DATETIMEOFFSET**

+ ***SQL Server affichera alors la chaîne de sortie avec les trois composants séparés par des espaces.***

####**4. Modification des décalages**

+ ***La fonction `SWITCHOFFSET()` nous permet de changer le fuseau horaire d'une chaîne d'entrée donnée. Si vous transmettez `un DATETIME ou un DATETIME2, SWITCHOFFSET()` suppose que vous êtes en `UTC`. Vous pouvez également passer un `DATETIMEOFFSET` pour passer d'un fuseau horaire connu à un autre. Ici, nous avons la date de la diapositive précédente. Cette date est dans un fuseau horaire qui est 4 heures plus lent que l'UTC, en particulier l'heure avancée de l'Est. Si nous voulons changer cela à l'heure de Los Angeles, en Californie, nous devrons savoir que Los Angeles était à l'heure avancée du Pacifique le 10 avril 2019. L'heure avancée du Pacifique est UTC -7. Ce résultat nous indique l'heure qu'il était à Los Angeles alors qu'il était 12h59 dans l'est des États-Unis : il était 9h59.***

####**5. Conversion en DATETIMEOFFSET**

+ ***Nous pouvons certainement utiliser la fonction `SWITCHOFFSET()` pour convertir une heure basée sur UTC d'un type de données `DATETIME2 en un DATETIMEOFFSET `mais il existe un moyen plus simple si nous avons juste besoin de donner à la date et à l'heure un décalage par rapport à UTC, et c'est la fonction `TODATETIMEOFFSET()`. Cette fonction prend deux paramètres : une date d'entrée et un fuseau horaire. À partir de là, il génère un `DATETIMEOFFSET`. Ici, j'ai la date que je sais être en heure avancée de l'Est mais qui a été stockée en tant que type de données `DATETIME2`. Maintenant, je l'ai en tant que `DATETIMEOFFSET`, ce qui me permettra de comparer les dates et les heures d'autres machines qui pourraient utiliser d'autres fuseaux horaires.***

####**6. Changement de fuseau horaire avec TODATETIMEOFFSET**

+ ***`TODATETIMEOFFSET()` peut également nous aider dans des scénarios un peu plus compliqués. Par exemple, supposons que nous ayons une date. Nous savons que le serveur enregistre les données à l'heure avancée du centre, ou UTC -5. Nous souhaitons créer un `DATETIMEOFFSET`, mais modifier le décalage pour l'aligner sur notre siège social à Bonn, en Allemagne. Tout d'abord, nous devons ajouter 7 heures à notre date d'entrée car il y a une différence de 7 heures entre UTC -5 et UTC +2. Maintenant que nous avons le bon moment, nous pouvons dire à SQL Server que la date a un décalage de plus 2 heures. Cela nous donne un résultat de 9h28 en heure d'été d'Europe centrale.***

Pour l'instant, sachez que `PERCENTILE_CONT()` prend un paramètre, le centile (une décimale allant de 0. à 1.). Le centile doit se trouver dans un groupe ordonné à l'intérieur de la clause `WITHIN GROUP et OVER` dans une certaine plage si vous devez partitionner les données. ```python
SELECT DISTINCT
	it.IncidentType,
	AVG(CAST(ir.NumberOfIncidents AS DECIMAL(4,2)))
	    OVER(PARTITION BY it.IncidentType) AS MeanNumberOfIncidents,
    --- Fill in the missing value
	PERCENTILE_CONT(0.5)
    	-- Inside our group, order by number of incidents DESC
    	WITHIN GROUP (ORDER BY ir.NumberOfIncidents DESC)
        -- Do this for each IncidentType value
        OVER (PARTITION BY it.IncidentType) AS MedianNumberOfIncidents,
	COUNT(1) OVER (PARTITION BY it.IncidentType) AS NumberOfRows
FROM dbo.IncidentRollup ir
	INNER JOIN dbo.IncidentType it
		ON ir.IncidentTypeID = it.IncidentTypeID
	INNER JOIN dbo.Calendar c
		ON ir.IncidentDate = c.Date
WHERE
	c.CalendarQuarter = 2
	AND c.CalendarYear = 2020;
```

####**5. Qu'en est-il du suréchantillonnage ?**

+ ***Jusqu'à présent, je me suis concentré sur le sous-échantillonnage car il est de loin le plus utile des deux. Le suréchantillonnage est le contraire du sous-échantillonnage : prendre des données agrégées et tenter de les désagréger. Le sous-échantillonnage fonctionne généralement en additionnant ou en comptant les résultats dans notre tableau, il est donc assez facile à faire. Le suréchantillonnage, quant à lui, nécessite une règle d'allocation : si vous avez des données à un grain horaire, comment allez-vous les allouer à chacune des 60 minutes ? Cela conduit au plus gros problème du suréchantillonnage : il fournit une granularité artificielle. Avec nos données horaires, en attribuant des enregistrements à des minutes particulières, nous devons prétendre que nous en savons plus que ce que nous savons réellement. Pourtant, il existe des utilisations pour le suréchantillonnage des données. Par exemple, avec des données au grain journalier, nous pourrions être en mesure d'utiliser un poisson ou un hypergéométrique pour modéliser les occurrences et générer des données de test au grain inférieur. Nous pouvons également utiliser des données suréchantillonnées pour avoir une idée de ce à quoi pourrait ressembler une minute ou une heure normale compte tenu des données quotidiennes, bien que trouver une réponse meilleure qu'une distribution uniforme nécessitera une certaine connaissance de la distribution sous-jacente des événements. Le sous-échantillonnage est beaucoup plus courant et beaucoup plus acceptable dans le monde des affaires que le suréchantillonnage - car chaque fois que j'ai eu besoin de suréchantillonner des données, je sous-échantillonne, sans exagération, des centaines de fois.***

####**2. Rollups hiérarchiques avec `ROLLUP`**

+ ***Le premier opérateur est `ROLLUP`, qui fonctionne mieux avec des données hiérarchiques. Ce code SQL présenté effectue une agrégation groupée sur une table appelée Table et calcule la somme des événements de chaque jour, regroupés par mois. Il ajoute ensuite une clause `WITH ROLLUP` pour ajouter une ligne de sous-total pour chaque mois distinct dans le jeu de résultats. `WITH ROLLUP` vient après `GROUP BY`. `ROLLUP` prendra chaque combinaison de la première colonne - mois dans ce cas - suivi de chaque valeur correspondante dans la deuxième colonne, et ainsi de suite, montrant nos agrégats pour chacun. `ROLLUP` est idéal pour un résumé des données hiérarchiques, mais si vous ne l'avez pas, vous pouvez utiliser l'un des autres opérateurs de regroupement.***

####**3. Agrégation cartésienne avec CUBE**

+ ***Pour les cas où nous voulons voir la combinaison complète de toutes les agrégations entre les colonnes, `CUBE` est à notre disposition. L'opérateur `CUBE` fonctionne exactement comme `ROLLUP`, glissant juste après la clause `GROUP BY`. `WITH CUBE` spécifie que le jeu de résultats doit inclure toutes les combinaisons possibles des colonnes répertoriées dans la clause `GROUP BY`. Sur la droite, nous pouvons voir un échantillon des lignes renvoyées par une opération `CUBE`. Dans tout scénario réaliste, nous obtiendrons probablement beaucoup plus de résultats que nous ne le souhaitons réellement. C'est là que la troisième option entre en jeu.***

####**4. Définir des ensembles de regroupement avec GROUPING SETS**

+ ***Avec `GROUPING SETS`, nous contrôlons les niveaux d'agrégation et pouvons inclure toute combinaison d'agrégats dont nous avons besoin. Par exemple, dans cette requête, nous définissons deux ensembles de regroupement : un sur la combinaison du type d'incident et du bureau, et un ensemble de regroupement vide pour nous donner le total général. Il en résulte une ligne avec le total général suivi de chacune des combinaisons spécifiques de type d'incident et de bureau. Si nous voulons ensuite inclure des agrégats distincts comme tous les types d'incidents ventilés quel que soit le bureau, nous pouvons les ajouter en tant qu'ensembles de regroupement supplémentaires. Vous pouvez créer n'importe quelle opération `ROLLUP ou CUBE` avec une série de `GROUPING SETS`, mais pour un plus grand nombre de colonnes, il est nettement plus rapide d'écrire `WITH ROLLUP` que de spécifier tous les ensembles de regroupement nécessaires pour reproduire son comportement.***

####**2. Fonctions de classement**

+ ***Il existe quatre fonctions de fenêtre de classement prises en charge par SQL Server : `ROW_NUMBER(), RANK(), DENSE_RANK() et NTILE()`. Nous couvrirons les trois premiers. `ROW_NUMBER()` fournit une valeur entière unique et ascendante garantie qui commence à 1 et continue jusqu'à la fin de l'ensemble. `RANK()` fournit une valeur entière croissante qui commence à 1, mais il n'est pas garanti qu'elle soit unique. Au lieu de cela, tous les liens dans la fenêtre obtiendront la même valeur, puis la valeur suivante obtiendra sa valeur `ROW_NUMBER()`. Si, par exemple, les deuxième et troisième entrées sont à égalité, les deux obtiendront un rang de 2 et la quatrième entrée obtiendra un rang de 4. `DENSE_RANK()` se comporte comme `RANK()`, sauf qu'il ne saute pas de nombres même avec des égalités. Si les deuxième et troisième entrées sont à égalité, les deux obtiendront un rang dense de 2 et la quatrième entrée obtiendra un rang dense de 3.***

####**3. Calcul des numéros de ligne**

+ ***Toutes les fonctions de classement nécessitent une clause `OVER()` avec une clause `ORDER BY` à l'intérieur. Ici, nous classons nos résultats par le nombre de points marqués par ordre décroissant. Ce que nous récupérons est un entier unique et croissant commençant à 1.***

####**4. Calcul des rangs et des rangs denses**

+ ***Toutes les fonctions de classement ont la même syntaxe. Ici, vous pouvez voir `RANK() et DENSE_RANK()` ensemble, ressemblant à `ROW_NUMBER()`. La différence réside dans les résultats : alors que `ROW_NUMBER()` nous a donné un entier unique par enregistrement, nous obtenons quatre groupes de résultats. La différence entre `RANK() et DENSE_RANK()` est le comportement des liens.***

####**5. Cloisons**

+ ***En plus d'une clause `ORDER BY`, la clause `OVER()` dans une fonction de fenêtre peut accepter une clause `PARTITION BY` qui divise la fenêtre par une colonne ou un ensemble de colonnes. Nous allons maintenant présenter les deux équipes qui ont joué et répartir les points marqués par équipe. `SQL Server` gère cela et nous fournit un numéro de ligne unique et croissant pour chaque équipe, de sorte que l'équipe de l'Arizona a trois entrées et l'équipe de Floride a trois entrées. Notez que les numéros de ligne de Florida sont réinitialisés à 1 - nous sommes assurés de commencer à 1 pour chaque fenêtre et la clause `PARTITION BY` définit notre fenêtre.***

####**6. Fonctions d'agrégation**

+ ***En plus des fonctions de classement, nous pouvons également utiliser des fenêtres sur les fonctions d'agrégation. Pour rappel, les fonctions d'agrégation incluent des fonctions telles que `AVG(), COUNT(), MAX(), MIN() et SUM()`, ainsi que plusieurs autres. Ici, nous voulons voir le nombre maximum de points marqués par chaque équipe ainsi que les points qu'ils ont marqués dans un match particulier. Au lieu d'écrire deux requêtes distinctes, nous pouvons obtenir les deux informations dans une seule requête. Cela nous permet de voir à quel point chaque équipe est proche de son efficacité offensive maximale.***

####**7. Agrégations avec fenêtres vides**

+ ***En allant un peu plus loin, une fonction d'agrégation avec une clause `OVER()` vide fait la même chose que la fonction d'agrégation non fenêtrée à une différence près : elle ne nécessite pas de grouper par colonnes non agrégées. Voici un exemple où nous utilisons la fonction `MAX()` avec une clause `OVER()` vide. Ce que nous récupérons, c'est chaque ligne ainsi que le nombre maximum de points marqués. Cela revient à écrire une sous-requête ou une expression de table commune qui interroge les données deux fois, une fois pour obtenir les lignes individuelles et une seconde fois pour obtenir le nombre maximum d'exécutions.***

+ *Dans cet exercice, nous aimerions déterminer la fréquence à laquelle nous voyons chaque incident de type 3 dans notre ensemble de données. Nous aimerions classer le nombre d'incidents par ordre décroissant , de sorte que la date avec le plus grand nombre d'incidents ait un numéro de ligne, un rang et un rang dense de 1, et ainsi de suite. ```python
SELECT
	ir.IncidentDate,
	ir.NumberOfIncidents,
    -- Fill in each window function and ordering
    -- Note that all of these are in descending order!
	ROW_NUMBER() OVER (ORDER BY ir.NumberOfIncidents DESC) AS rownum,
	RANK() OVER (ORDER BY ir.NumberOfIncidents DESC) AS rk,
	DENSE_RANK() OVER (ORDER BY ir.NumberOfIncidents DESC) AS dr
FROM dbo.IncidentRollup ir
WHERE
	ir.IncidentTypeID = 3
	AND ir.NumberOfIncidents >= 8
ORDER BY
	ir.NumberOfIncidents DESC;
```

####**Voir les périodes antérieures et futures**

+ ***Les fonctions `LAG()et LEAD()` window nous donnent la possibilité de regarder en arrière ou en avant dans le temps, respectivement. Cela nous donne la possibilité de comparer les données d'une période à l'autre en une seule requête simple.***

+ *Dans cet exercice, nous voulons comparer le nombre d'incidents de sécurité par jour pour les types d'incidents 1 et 2 en juillet 2019, en particulier la période commençant le 2 juillet et se terminant le 31 juillet.*

####**Instructions**

+ Remplissez la fonction de fenêtre pour renvoyer le nombre d'incidents du jour précédent, partitionné par ID de type d'incident et classé par date d'incident.
+ Remplissez la fonction de fenêtre pour renvoyer le nombre d'incidents du jour suivant, partitionné par ID de type d'incident et classé par date d'incident.

```python
SELECT
	ir.IncidentDate,
	ir.IncidentTypeID,
    -- Get the prior day's number of incidents
	LAG(ir.NumberOfIncidents, 1) OVER (
      	-- Partition by incident type ID
		PARTITION BY ir.IncidentTypeID
      	-- Order by incident date
		ORDER BY ir.IncidentDate
	) AS PriorDayIncidents,
	ir.NumberOfIncidents AS CurrentDayIncidents,
    -- Get the next day's number of incidents
	LEAD(ir.NumberOfIncidents, 1) OVER (
      	-- Partition by incident type ID
		PARTITION BY ir.IncidentTypeID
      	-- Order by incident date
		ORDER BY ir.IncidentDate
	) AS NextDayIncidents
FROM dbo.IncidentRollup ir
WHERE
	ir.IncidentDate >= '2019-07-02'
	AND ir.IncidentDate <= '2019-07-31'
	AND ir.IncidentTypeID IN (1, 2)
ORDER BY
	ir.IncidentTypeID,
	ir.IncidentDate;
```

####**Voir les trois périodes précédentes**

Cela nous donne la possibilité de comparer les données d'une période à l'autre en une seule requête simple. Chaque appel à `LAG() ou LEAD()` renvoie soit une valeur `NULL`, soit une seule ligne. + *Dans cet exercice, nous voulons comparer le nombre d'incidents de sécurité par jour pour les types d'incidents 1 et 2 en juillet 2019, en particulier la période commençant le 2 juillet et se terminant le 31 juillet. + Remplissez le SQL pour renvoyer le nombre d'incidents de la période précédente.
+ Remplissez le SQL pour renvoyer le nombre d'incidents de la période suivante.

```python
SELECT
	ir.IncidentDate,
	ir.IncidentTypeID,
    -- Fill in two periods ago
	LAG(ir.NumberOfIncidents, 2) OVER (
		PARTITION BY ir.IncidentTypeID
		ORDER BY ir.IncidentDate
	) AS Trailing2Day,
    -- Fill in one period ago
	LAG(ir.NumberOfIncidents, 1) OVER (
		PARTITION BY ir.IncidentTypeID
		ORDER BY ir.IncidentDate
	) AS Trailing1Day,
	ir.NumberOfIncidents AS CurrentDayIncidents,
    -- Fill in next period
	LEAD(ir.NumberOfIncidents, 1) OVER (
		PARTITION BY ir.IncidentTypeID
		ORDER BY ir.IncidentDate
	) AS NextDay
FROM dbo.IncidentRollup ir
WHERE
	ir.IncidentDate >= '2019-07-01'
	AND ir.IncidentDate <= '2019-07-31'
	AND ir.IncidentTypeID IN (1, 2)
ORDER BY
	ir.IncidentTypeID,
	ir.IncidentDate;
```

####**Calcul des jours écoulés entre les incidents**

Quelqu'un de la direction l'a également remarqué et, fin juillet, a voulu connaître le nombre de jours entre les incidents. Pour ce faire, nous allons calculer deux valeurs : le nombre de jours depuis l'incident précédent et le nombre de jours jusqu'au prochain incident.***

+ *Rappel qui `DATEDIFF()` donne la différence entre deux dates. Nous pouvons combiner cela avec `LAG() et LEAD()` pour obtenir nos résultats.*

####**Instructions**

+ Calculez le nombre de jours depuis le dernier incident en utilisant une combinaison de `DATEDIFF() et LAG() ou LEAD()`.
+ Calculez les jours jusqu'au prochain incident en utilisant une combinaison de `DATEDIFF() et LAG() ou LEAD()`.
+ `REMARQUE` : vous n'aurez pas besoin d'utiliser la colonne NumberOfIncidents dans cet exercice.

```python
SELECT
	ir.IncidentDate,
	ir.IncidentTypeID,
    -- Fill in the days since last incident
	DATEDIFF(DAY, LAG(ir.IncidentDate, 1) OVER (
		PARTITION BY ir.IncidentTypeID
		ORDER BY ir.IncidentDate
	), ir.IncidentDate) AS DaysSinceLastIncident,
    -- Fill in the days until next incident
	DATEDIFF(DAY, ir.IncidentDate, LEAD(ir.IncidentDate, 1) OVER (
		PARTITION BY ir.IncidentTypeID
		ORDER BY ir.IncidentDate
	)) AS DaysUntilNextIncident
FROM dbo.IncidentRollup ir
WHERE
	ir.IncidentDate >= '2019-07-02'
	AND ir.IncidentDate <= '2019-07-31'
	AND ir.IncidentTypeID IN (1, 2)
ORDER BY
	ir.IncidentTypeID,
	ir.IncidentDate;
```

```python
-- This section focuses on entrances:  CustomerVisitStart
SELECT
	dsv.CustomerID,
	dsv.CustomerVisitStart AS TimeUTC,
	1 AS EntryCount,
    -- We want to know each customer's entrance stream
    -- Get a unique, ascending row number
	ROW_NUMBER() OVER (
      -- Break this out by customer ID
      PARTITION BY dsv.CustomerID
      -- Ordered by the customer visit start date
      ORDER BY dsv.	CustomerVisitStart
    ) AS StartOrdinal
FROM dbo.DaySpaVisit dsv
UNION ALL
-- This section focuses on departures:  CustomerVisitEnd
SELECT
	dsv.CustomerID,
	dsv.CustomerVisitEnd AS TimeUTC,
	-1 AS EntryCount,
	NULL AS StartOrdinal
FROM dbo.DaySpaVisit dsv
```

+ *Les résultats de l'exercice précédent se trouvent maintenant dans une table temporaire appelée #StartStopPoints. Les colonnes de ce tableau sont CustomerID, TimeUTC, EntryCount et StartOrdinal. Ce sont les seules colonnes que vous devrez utiliser dans cet exercice. TimeUTCreprésente l'heure de l'événement, EntryCount indique le changement net pour l'événement (+1 ou -1), et StartOrdinal apparaît pour les événements d'entrée et donne l'ordre d'entrée.*

####**Instructions**

+ Remplissez la fonction de fenêtre appropriée `(ROW_NUMBER())` pour créer un flux d'arrivées et de départs dans l'ordre chronologique.
+ Partition par l'ID client pour calculer un résultat par utilisateur.
+ Triez par l'heure de l'événement et résolvez les égalités en utilisant la valeur ordinale de début.

```python
SELECT s.*,
    -- Build a stream of all check-in and check-out events
	ROW_NUMBER() OVER (
      -- Break this out by customer ID
      PARTITION BY s.CustomerID
      -- Order by event time and then the start ordinal
      -- value (in case of exact time matches)
      ORDER BY s.TimeUTC, s.StartOrdinal
    ) AS StartOrEndOrdinal
FROM #StartStopPoints s;
```

#####**Compléter l'analyse de la fraude**

+ ***Jusqu'à présent, nous avons divisé les données du spa de jour en un flux d'entrées et de sorties et classé ce flux par ordre chronologique. Ce flux contient deux champs critiques StartOrdinal et StartOrEndOrdinal. StartOrdinal est l'ordre chronologique de toutes les entrées. StartOrEndOrdinalcontient toutes les entrées et sorties dans l'ordre. Armés de ces deux informations, nous pouvons trouver le nombre maximum de visites simultanées.**

+ *Les résultats de l'exercice précédent se trouvent maintenant dans une table temporaire appelée #StartStopOrder.*

####**Consignes**
+ Remplissez la clause `HAVING` pour déterminer les cas avec plus de 2 visiteurs simultanés.
+ Remplissez la clause `ORDER BY` pour montrer à la direction les pires contrevenants : ceux qui ont les valeurs les plus élevées pour `MaxConcurrentCustomerVisits`.

```python
SELECT
	s.CustomerID,
	MAX(2 * s.StartOrdinal - s.StartOrEndOrdinal) AS MaxConcurrentCustomerVisits
FROM #StartStopOrder s
WHERE s.EntryCount = 1
GROUP BY s.CustomerID
-- The difference between 2 * start ordinal and the start/end
-- ordinal represents the number of concurrent visits
HAVING MAX(2 * s.StartOrdinal - s.StartOrEndOrdinal) > 2
-- Sort by the largest number of max concurrent customer visits
ORDER BY MaxConcurrentCustomerVisits DESC;
```
