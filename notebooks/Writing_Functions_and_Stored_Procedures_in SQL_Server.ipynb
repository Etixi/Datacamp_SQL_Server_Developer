+ ***Il est temps pour vous de faire de l'EDA temporelle sur le jeu de données BikeShare. Rédigez une requête pour déterminer le nombre de transactions existantes par jour.***

+ *Parfois, les jeux de données ont plusieurs sources et cette requête peut vous aider à comprendre s'il vous manque des données.*

####**Instructions**

+ Utilisez CONVERT() pour SELECT et GROUP BY la partie date de StartDate.
+ Utilisez COUNT() pour mesurer le nombre d'enregistrements existants pour chaque StartDate.
+ Triez les résultats par la partie date de StartDate.

```python
SELECT
	-- Count the number of IDs
	COUNT(ID) AS Count,
    -- Use DATEPART() to evaluate the SECOND part of StartDate
    "StartDate" = CASE WHEN DATEPART(SECOND, StartDate) = 0 THEN 'SECONDS = 0'
					   WHEN DATEPART(SECOND, StartDate) > 0 THEN 'SECONDS > 0' END
FROM CapitalBikeShare
GROUP BY
    -- Use DATEPART() to Group By the CASE statement
	CASE WHEN DATEPART(SECOND, StartDate) = 0 THEN 'SECONDS = 0'
		 WHEN DATEPART(SECOND, StartDate) > 0 THEN 'SECONDS > 0' END
```

***Vous pouvez maintenant utiliser `CASE()` et `DATEPART()` pour classer par les parties de date d'un datetime.***

####**Quel jour de la semaine est le plus chargé ?**

+ ***Maintenant que nous avons vérifié qu'il y a des secondes de manière cohérente dans notre ensemble de données, nous pouvons calculer la durée totale du trajet pour chaque jour de la semaine.***

####**Instructions**

+ Utilisez `DATENAME()` pour `SÉLECTIONNER` la valeur du jour de la semaine pour StartDate.
+ Utilisez `SUM()` et `DATEDIFF()` pour calculer TotalTripHours. (commençant par secondes).
+ Regroupez par le résultat `DATENAME()` et récapitulez TotalTripHours.
+ Triez TotalTripHours par ordre décroissant.

```python
SELECT
    -- Select the day of week value for StartDate
	DATENAME(weekday, StartDate) as DayOfWeek,
    -- Calculate TotalTripHours
	SUM(DATEDIFF(SECOND, StartDate, EndDate))/ 3600 as TotalTripHours
FROM CapitalBikeShare
-- Group by the day of week
GROUP BY DATENAME(weekday, StartDate)
-- Order TotalTripHours in descending order
ORDER BY TotalTripHours DESC
```

+ Utilisez `DATENAME()` et `CONVERT()` pour sélectionner le `WEEKDAY`.
+ Utilisez `WHERE` pour inclure uniquement les samedis.

```python
SELECT
	-- Calculate TotalRideHours using SUM() and DATEDIFF()
  	SUM(DATEDIFF(SECOND, StartDate, EndDate))/ 3600 AS TotalRideHours,
    -- Select the DATE portion of StartDate
  	CONVERT(DATE, StartDate) AS DateOnly,
    -- Select the WEEKDAY
  	DATENAME(WEEKDAY, CONVERT(DATE, StartDate)) AS DayOfWeek
FROM CapitalBikeShare
-- Only include Saturday
WHERE DATENAME(WEEKDAY, StartDate) = 'Saturday'
GROUP BY CONVERT(DATE, StartDate);
```

####**4. CASTing**
+ Comment pouvons-nous combiner une variable de date et d’heure en une seule variable datetime ? Nous pouvons utiliser la fonction `CAST()`.

+ **`CAST()` est très similaire à la fonction `CONVERT()` que vous avez apprise précédemment, bien qu’il s’agisse de la norme `ANSI` et qu’elle fonctionne sur toutes les plates-formes `SQL`**.

+ **`CAST()` convertit une expression en un autre type de données. Considérez `CAST()` comme le transfert du contenu d’un casier vers un casier de taille différente. L’opérateur add peut être utilisé avec `CAST()` pour combiner deux variables datetime, mais d’abord, nous devons `@StartTime CAST()` d’une heure à une heure et d’une heure de date, ainsi que `@BeginDate` d’une date à une dateTime. SQL utilise des règles de valeur par défaut lors de l’exécution d’instructions `CAST()` afin que vous puissiez `CAST()` une heure à une datetime même en l’absence de valeur date. Lorsqu’une variable de temps est convertie d’un casier de temps à un casier de date, SQL en fait une date et une heure valide et inclut une valeur de date par défaut.**

+ Utilisez une instruction select pour additionner la différence entre `StartDate et EndDate` des transactions.
+ N'incluez que les transactions dont la StartDate est supérieure à `@StartDateParm` et inférieure à `@EndDateParm`.

```python
-- Create the function
CREATE FUNCTION SumRideHrsDateRange (@StartDateParm datetime, @EndDateParm datetime)
-- Specify return data type
RETURNS numeric
AS
BEGIN
RETURN
-- Sum the difference between StartDate and EndDate
(SELECT SUM(DATEDIFF(second, StartDate, EndDate))/3600
FROM CapitalBikeShare
-- Include only the relevant transactions
WHERE StartDate > @StartDateParm and StartDate < @EndDateParm)
END
```

####**2. Fonctions tables en ligne (ITVF)**
+ Il existe deux types différents de fonctions table : inline et multi instruction.

+ Passons en revue le code d’une fonction de table en ligne, ou `ITVF`.

+ La première ligne est similaire à une fonction scalaire avec les mots-clés `CREATE FUNCTION`, suivis du nom de la fonction, puis du nom du paramètre d’entrée et du type de données.

+ L’affectation de valeurs par défaut aux paramètres est une option dans toutes les fonctions définies par l’utilisateur.

+ La ligne suivante contient le mot clé `RETURNS`, mais `TABLE` suit au lieu d’un type de données scalaire. Le mot-clé `RETURN` est suivi de l’instruction `SELECT`.

+ Aucun bloc `BEGIN END` n’est nécessaire car `SQL Server` renvoie les résultats de l’instruction `SELECT` unique.

+ Les fonctions scalaires nécessitent l’utilisation du bloc `BEGIN END`, qu’il s’agisse ou non d’une instruction unique, mais une fonction table ne nécessite pas `BEGIN END` si le corps de la fonction est une instruction unique.

+ Les noms de colonne doivent être affectés dans l’instruction `SELECT` car une table est renvoyée.

+ Cette fonction renvoie un tableau contenant le nombre de trajets et la distance totale parcourue pour chaque PickupLocation où la date de début de la transaction est égale à la valeur de paramètre transmise.

####**3. Fonction table multi-instructions (MSTVF)**


+ Passons maintenant en revue une fonction à valeurs de table multi-instructions, ou MSTVF.

+ Encore une fois, la première ligne est similaire à `INLINE et SCALAR` avec les mots-clés `CREATE FUNCTION`, le nom de la fonction et les définitions de paramètres d’entrée.

+ Le mot-clé `RETURNS` sur la deuxième ligne est suivi d’une définition de variable `TABLE`, y compris les noms de colonne et les types de données pour chacun.

+ Comme il s’agit d’une instruction multiple, nous devons utiliser un bloc `BEGIN END` pour contenir les instructions SQL multiples.

+ Nous renvoyant la variable de table, nous devons donc `INSERT` les résultats de l’instruction `SELECT`. Après `INSERT et SELECT`, nous devons indiquer à la fonction de `RETURN` la variable de table. `RETURN` est la dernière instruction du bloc `BEGIN END`. N’oubliez pas d’affecter le type de données correct aux colonnes de la variable de table. Ils doivent s’aligner sur les données qui seront insérées.

####**4. Différences - ITVF vs MSTVF**
+ Quelles sont les principales différences entre les fonctions inline et multi-instructions Table Valued Functions ?

+ Inline renvoie les résultats d’une instruction SELECT ainsi que les noms de colonne affectés.

+ Il ne contient pas de variable de table, de bloc `BEGIN END ou INSERT`.

+ Il n’est probablement pas surprenant qu’il fonctionne généralement plus rapidement par rapport à une instruction multiple.

+ Cela devrait toujours être votre itinéraire préféré. Si vous pouvez écrire votre requête dans une seule instruction, vous devriez le faire.

+ Une fonction table à valeurs multi-instructions doit `DECLARE` une variable de table à renvoyer, contenir un bloc `BEGIN END, INSERT` data dans la variable table et `RETURN` doit être la dernière instruction du bloc `BEGIN END`. Ce type de fonction table ne doit être utilisé que lorsque vous ne pouvez pas écrire la requête dans une seule instruction.

####**TVF multi-instructions**

+ Utilisez `CAST` pour sélectionner et regrouper par `StartDate` en tant que type de données de date.

```python
-- Create the function
CREATE FUNCTION  CountTripAvgDuration (@Month CHAR(2), @Year CHAR(4))
-- Specify return variable
RETURNS @DailyTripStats TABLE(
	TripDate	date,
	TripCount	int,
	AvgDuration	numeric)
AS
BEGIN
-- Insert query results into @DailyTripStats
INSERT INTO @DailyTripStats
SELECT
    -- Cast StartDate as a date
	CAST(StartDate AS date),
    COUNT(ID),
    AVG(Duration)
FROM CapitalBikeShare
WHERE
	DATEPART(month, StartDate) = @Month AND
    DATEPART(year, StartDate) = @Year
-- Group by StartDate as a date
GROUP BY CAST(StartDate AS date)
-- Return
RETURN
END
```

###**Maintien des fonctions définies par l’utilisateur**
####**1. Maintien des fonctions définies par l’utilisateur**

+ ***Je sais que la maintenance ne semble pas aussi excitante que la création et l’exécution, mais si vous ne prenez pas bien soin de votre travail acharné, qui le fera?***

####**2. Fonction `ALTER`**
+ Vous vous êtes peut-être déjà demandé, comment puis-je changer une fonction qui a déjà été créée? Le mot-clé `ALTER` peut être utilisé pour cela.

+ Ici, nous modifions la fonction `SumLocationStats()` que nous avons créée précédemment.

+ Le nom du paramètre d’entrée est maintenant `@EndDate` au lieu de `@BeginDate` et est comparé à la colonne `DropOffDate` au lieu de PickupDate.

####**3. CRÉER OU MODIFIER**
+ Vous pouvez également utiliser des mots-clés `CREATE OR ALTER` ensemble.

+ Ceci est utile pendant le processus de développement lorsque vous apportez de nombreuses modifications ultérieures. `SQL Server` ne vous permet pas de CRÉER une fonction qui existe déjà, mais `CREATE OR ALTER` s’exécutera sans problème.

####**4. Fonction `DROP`**
+ Il existe certaines limitations à `ALTER`.

+ Si vous souhaitez remplacer une fonction table d’instruction multiple par Inline ou vice versa, vous ne pouvez pas utiliser `ALTER`.

+ Au lieu de cela, pour effectuer ce type de changement structurel, vous devez utiliser `DROP` pour supprimer la fonction. Après sa suppression, utilisez CREATE comme instruction `Inline ou Multi`.

+ `DROP` doit être utilisé chaque fois que vous devez supprimer une fonction.

+ Assurez-vous de garder votre travail bien rangé et supprimez les fonctions que vous avez créées mais que vous ne prévoyez pas d’utiliser à l’avenir.

+ `DROP` s’exécutera tant que l’utilisateur de votre base de données aura l’autorisation de le faire.

####**5. Le déterminisme améliore les performances**
+ Le déterminisme est une caractéristique d’une fonction et est vrai lorsqu’une fonction renvoie le même résultat avec les mêmes paramètres d’entrée et le même état de la base de données.

+ Une fonction est non déterministe si elle peut renvoyer une valeur différente avec les mêmes paramètres d’entrée et l’état de la base de données.

####**6. Le déterminisme améliore les performances**

+ `GetRideHoursOneDay()` est un exemple de fonction déterministe tandis que `GetTomorrow()` est non déterministe. `GetTomorrow()`, tout comme `GETDATE()`, ne renvoie pas le même résultat car il est toujours par rapport à la date actuelle.

+ Parfois, les fonctions ne peuvent pas être rendues déterministes, mais nous devrions essayer de les rendre déterministes si possible.

+ Lorsqu’une fonction est déterministe, `SQL Server` peut indexer les résultats, ce qui contribue à améliorer les performances.

+ Le déterminisme des fonctions SQL intégrées ne peut pas être ajusté.

+ Lorsque vous utilisez des `fonctions SQL natives`, vous devez vérifier si elles sont déterministes ou non.

+ L’un des aspects du déterminisme consiste à renvoyer le même résultat basé sur le même état de la base de données.

+ Comment une fonction peut-elle vérifier que l’état de la base de données ne changera pas d’une manière qui affecterait son résultat ? L’option schemabinding de la fonction peut être activée.

####**7. Liaison de schémas**
+ **Un schéma est une collection d’objets de base de données associés à un propriétaire de base de données particulier tel que `dbo`**.

+ Les fonctions définies par l’utilisateur peuvent référencer des objets de base de données tels que des tables, des colonnes, des types de données ou d’autres fonctions.

+ Si l’option `SCHEMABINDING` est activée dans la fonction, `SQL Server` empêche les modifications apportées aux objets de base de données qu’il référence.

+ Par exemple, si quelqu’un tentait de renommer la colonne `PickupDate` de la table `YellowTripData`, `SQL` empêcherait cette modification car la fonction `GetRideHrsOneDay()` est liée à son schéma. Pour ce faire, nous incluons les mots-clés `WITH SCHEMABINDING` après la clause `RETURNS`. ###**Procédures stockées**
####**1. Procédures stockées**

+ ***Il est temps d’en apprendre davantage sur les procédures stockées, un autre excellent outil à ajouter à vos compétences SQL.***

####**2. Qu’est-ce qu’une procédure stockée ?**
+ **Qu’est-ce qu’une procédure stockée, ou SP ?**

  + **C’est une routine qui peut accepter des paramètres d’entrée, effectuer une action et renvoyer des paramètres d’état de réussite et de sortie**.
  
  + **Certaines de ces caractéristiques devraient vous sembler familières; Peut-être comme les fonctions définies par l’utilisateur dont nous avons parlé plus tôt? Ils sont similaires, mais pas les mêmes.**
  
  + Dans les procédures stockées, vous pouvez `EXÉCUTER INSERT, UPDATE, SELECT, DELETE et appeler d’autres procédures stockées`. Ces instructions peuvent modifier des données en dehors de la portée `SP`.

####**3. Pourquoi utiliser des procédures stockées ?**

+ ***Pourquoi devrions-nous utiliser des procédures stockées ?***

+ **Ils peuvent réduire le temps d’exécution, le trafic réseau et permettre une programmation modulaire comme les fonctions définies par l’utilisateur, mais ils peuvent également améliorer la sécurité de la base de données**.

+ **Les procédures stockées peuvent être écrites et utilisées pour aider à prévenir les attaques par injection `SQL` qui sont une technique néfaste où des instructions `SQL` sont injectées dans une base de données.**

####**4. Quelle est la différence?**
+ Passons en revue les différences entre les fonctions définies par l’utilisateur et les procédures stockées.

+ Les seules caractéristiques qu’ils partagent réellement sont l’acceptation des paramètres d’entrée et l’exécution d’une action. Le reste sont des différences distinctes.

+ Les SP n’ont pas besoin de renvoyer une valeur et ne peuvent techniquement pas renvoyer un résultat de valeur de table.

+ **Ils ne peuvent pas être exécutés dans une instruction `SELECT` simple, mais peuvent renvoyer des paramètres de sortie, un état de retour et des jeux de résultats. Vous pouvez exécuter des fonctions et des `SP` à partir d’un `SP`, ainsi que des instructions `INSERT, UPDATE et DELETE`. Vous pouvez également gérer efficacement les erreurs dans les procédures stockées.**

####**5. CREATE PROCEDURE avec paramètre OUTPUT**

+ Examinons les quatre premières lignes de la syntaxe SQL Server pour créer une procédure stockée.

+ Ici, nous utilisons la logique de la `fonction GetRideHrsOneDay()` décrite précédemment. Nous utilisons le mot-clé `CREATE PROCEDURE` suivi du nom du schéma et de la procédure stockée.

+ Les noms de procédures stockées doivent être uniques dans l’ensemble du schéma, y compris les fonctions définies par l’utilisateur.

+ Vous ne pouvez pas avoir le même nom qu’une fonction définie par l’utilisateur, nous avons donc ajouté un préfixe cuspide ici.

+ Le nom du SP est suivi du paramètre d’entrée `@DateParm` et de son type de données de date. Un paramètre de sortie, `@RideHrsOut`, est défini comme un type de données numérique et le mot clé `OUTPUT` indique qu’il doit être renvoyé en sortie. Les paramètres doivent toujours être nommés avec le symbole arobase comme premier caractère, mais il n’est pas nécessaire qu’ils soient placés entre parenthèses, comme dans les fonctions définies par l’utilisateur.

####**6. CREATE PROCEDURE avec paramètre OUTPUT**
+ **L’instruction `SET NOCOUNT ON` de la 4e ligne empêche SQL de renvoyer à l’appelant le nombre de lignes affectées par la procédure stockée.**

+ **Ceci est facultatif, et certains considèrent qu’il est recommandé de le faire, mais cela peut entraîner des problèmes pour l’application appelante si elle s’attend à ce que ces données soient renvoyées.**

+ **Le reste de l’instruction est identique à la `FDU GetRideHrsOneDay()` créée précédemment, sauf que nous affectons le résultat de l’instruction `SELECT` au paramètre de sortie `@RideHrsOut`. Le mot-clé `RETURN` est facultatif et indique au `SP` de renvoyer immédiatement le paramètre de sortie à l’application appelante.**

####**7. Paramètres de sortie et valeurs de retour**

+ Que sont les paramètres de sortie et les valeurs de retour ?

+ Les paramètres de sortie sont définis de la même manière que les paramètres d’entrée, mais avec le mot-clé `OUTPUT`. Il peut s’agir de n’importe quel type de données, à l’exception de la valeur de table. Vous pouvez également déclarer plusieurs paramètres de sortie par procédure stockée.

+ Les valeurs renvoyées dans les SP sont généralement utilisées pour indiquer la réussite ou l’échec de l’application appelante.

+ Les `SP` ne peuvent renvoyer que des types de données entiers et zéro indique la réussite. Des valeurs différentes de zéro indiquent que le `SP` a rencontré un certain type d’erreur lors de l’exécution.

+ Créez une procédure stockée nommée `cuspSumRideHrsSingleDay` dans le schéma dbo qui accepte une date et renvoie le nombre total d'heures de trajet pour la date passée.

####**Instructions**

+ Créez une procédure stockée appelée `cuspSumRideHrsSingleDay` dans le schéma `dbo`.

+ N'envoyez pas le nombre de lignes à l'appelant.
+ Attribuez le résultat de la requête à `@RideHrsOut` et incluez le mot-clé `RETURN`.

```python
-- Create the stored procedure
CREATE PROCEDURE dbo.cuspSumRideHrsSingleDay
    -- Declare the input parameter
	@DateParm date,
    -- Declare the output parameter
	@RideHrsOut numeric OUTPUT
AS
-- Don't send the row count
SET NOCOUNT ON
BEGIN
-- Assign the query result to @RideHrsOut
SELECT
	@RideHrsOut = SUM(DATEDIFF(second, StartDate, EndDate))/3600
FROM CapitalBikeShare
-- Cast StartDate as date and compare with @DateParm
WHERE CAST(StartDate AS date) = @DateParm
RETURN
END
```

###**Oh `CRUD`!**
####**1. Oh `CRUD`!**

####**2. Pourquoi des procédures stockées pour `CRUD` ?**

+ Des procédures stockées existent dans la source de données et la couche d’accès aux données n’a besoin que de les exécuter, au lieu d’exécuter des requêtes SQL plus complexes. Cela dissocie le code SQL de la couche d’accès aux données.

+ **L’optimiseur de requêtes SQL Server crée un plan d’exécution lorsqu’une procédure stockée est exécutée pour la première fois**.

+ Le plan est une tentative de calculer le moyen le plus efficace d’implémenter la requête soumise.

+ Le moteur de base de données suit le plan d’exécution existant.

+ Étant donné qu’il faut beaucoup de ressources à l’optimiseur de requêtes pour créer le plan d’exécution, elles sont mises en cache pour une utilisation ultérieure.

+ Cette réutilisation du plan d’exécution permet aux procédures stockées de s’exécuter plus rapidement par rapport aux requêtes SQL ad hoc qui doivent être analysées par l’optimiseur de requêtes chaque fois qu’elles sont exécutées.

####**Utiliser SP pour METTRE À JOUR**
+ ***Créez une procédure stockée nommée `cuspRideSummaryUpdate` dans le schéma `dbo` qui mettra à jour un enregistrement existant dans la table `RideSummary`.***

####**Instructions**

+ Le `SP` doit accepter les paramètres d'entrée pour chaque colonne `RideSummary` et être nommé `@Date et @RideHrs`.
+ Faites du paramètre `@Date` un type de données de date et `@RideHrs` un type de données numérique.
+ Utilisez les mots-clés `UPDATE et SET` pour affecter les valeurs de paramètre à l'enregistrement `RideSummar` où `@Date` correspond à la valeur Date.

```python
-- Create the stored procedure
CREATE PROCEDURE dbo.cuspRideSummaryUpdate
	-- Specify @Date input parameter
	(@Date date,
     -- Specify @RideHrs input parameter
     @RideHrs numeric(18,0))
AS
BEGIN
SET NOCOUNT ON
-- Update RideSummary
UPDATE RideSummary
-- Set
SET
	Date = @Date,
    RideHours = @RideHrs
-- Include records where Date equals @Date
WHERE Date = @Date
END;
```

```python
-- Alter the stored procedure
CREATE OR ALTER PROCEDURE dbo.cuspRideSummaryDelete
	-- (Incorrectly) specify @DateParm
	@DateParm nvarchar(30),
    -- Specify @Error
	@Error nvarchar(max) = NULL OUTPUT
AS
SET NOCOUNT ON
BEGIN
  -- Start of the TRY block
  BEGIN TRY
  	  -- Delete
      DELETE FROM RideSummary
      WHERE Date = @DateParm
  -- End of the TRY block
  END TRY
  -- Start of the CATCH block
  BEGIN CATCH
		SET @Error =
		'Error_Number: '+ CAST(ERROR_NUMBER() AS VARCHAR) +
		'Error_Severity: '+ CAST(ERROR_SEVERITY() AS VARCHAR) +
		'Error_State: ' + CAST(ERROR_STATE() AS VARCHAR) +
		'Error_Message: ' + ERROR_MESSAGE() +
		'Error_Line: ' + CAST(ERROR_LINE() AS VARCHAR)
  -- End of the CATCH block
  END CATCH
END;
```

+ Mettez à jour les enregistrements dans YellowTripData où TripDistance est 0 et définissez-les sur @AvgTripDistance.

```python
-- Create the stored procedure
CREATE PROCEDURE dbo.cuspImputeTripDistanceMean
AS
BEGIN
-- Specify @AvgTripDistance variable
DECLARE @AvgTripDistance AS numeric (18,4)

-- Calculate the average trip distance
SELECT @AvgTripDistance = AVG(TripDistance)
FROM YellowTripData
-- Only include trip distances greater than 0
WHERE TripDistance > 0

-- Update the records where trip distance is 0
UPDATE YellowTripData
SET TripDistance =  @AvgTripDistance
WHERE TripDistance = 0
END;
```

```python
SELECT
    -- Select the pickup day of week
	DATENAME(weekday, PickupDate) as DayofWeek,
    -- Calculate TotalAmount per TripDistance
	CAST(AVG(TotalAmount/
            -- Select TripDistance if it's more than 0
			CASE WHEN TripDistance > 0 THEN TripDistance
                 -- Use GetTripDistanceHotDeck()
     			 ELSE dbo.GetTripDistanceHotDeck() END) as decimal(10,2)) as 'AvgFare'
FROM YellowTripData
GROUP BY DATENAME(weekday, PickupDate)
-- Order by the PickupDate day of week
ORDER BY
     CASE WHEN DATENAME(weekday, PickupDate) = 'Monday' THEN 1
        WHEN DATENAME(weekday, PickupDate) = 'Tuesday' THEN 2
         WHEN DATENAME(weekday, PickupDate) = 'Wednesday' THEN 3
         WHEN DATENAME(weekday, PickupDate) = 'Thursday' THEN 4
         WHEN DATENAME(weekday, PickupDate) = 'Friday' THEN 5
         WHEN DATENAME(weekday, PickupDate) = 'Saturday' THEN 6
         WHEN DATENAME(weekday, PickupDate) = 'Sunday' THEN 7
END ASC;
```

Après discussion avec les parties prenantes, vous devez omettre les enregistrements où le TripDistance est égal à zéro.***

####**Instructions**

+ Sélectionnez et regroupez par jour de semaine de ramassage et Borough.
+ Calculez AvgFarePerKM avec dbo.ConvertDollar() et dbo.ConvertMiletoKM()en utilisant le taux de change de l'euro .88.
+ Afficher AvgFarePerKM en devise allemande RideCount et TotalRideMin en nombres allemands.
+ Omettre les enregistrements où TripDistance est 0.

```python
CREATE OR ALTER PROCEDURE dbo.cuspBoroughRideStats
AS
BEGIN
SELECT
    -- Calculate the pickup weekday
	DATENAME(weekday, PickupDate) AS 'Weekday',
    -- Select the Borough
	Zone.Borough AS 'PickupBorough',
    -- Display AvgFarePerKM as German currency
	FORMAT(AVG(dbo.ConvertDollar(TotalAmount, .88)/dbo.ConvertMiletoKM(TripDistance)), 'c', 'de-de') AS 'AvgFarePerKM',
    -- Display RideCount in the German format
	FORMAT(COUNT(ID), 'n', 'de-de') AS 'RideCount',
    -- Display TotalRideMin in the German format
	FORMAT(SUM(DATEDIFF(SECOND, PickupDate, DropOffDate))/60, 'n', 'de-de') AS 'TotalRideMin'
FROM YellowTripData
INNER JOIN TaxiZoneLookup AS Zone
ON PULocationID = Zone.LocationID
-- Only include records where TripDistance is greater than 0
WHERE TripDistance> 0
-- Group by pickup weekday and Borough
GROUP BY DATENAME(WEEKDAY, PickupDate), Zone.Borough
ORDER BY CASE WHEN DATENAME(WEEKDAY, PickupDate) = 'Monday' THEN 1
	     	  WHEN DATENAME(WEEKDAY, PickupDate) = 'Tuesday' THEN 2
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Wednesday' THEN 3
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Thursday' THEN 4
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Friday' THEN 5
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Saturday' THEN 6
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Sunday' THEN 7 END,
		 SUM(DATEDIFF(SECOND, PickupDate, DropOffDate))/60
DESC
END;
```

####**Lieux de ramassage par quart de travail**

+ ***Il est temps de résoudre le deuxième objectif de l'analyse de rentabilisation. Quels sont les `AvgFarePerKM, RideCount et TotalRideMin` pour chaque lieu de prise en charge et quart de travail dans un arrondissement de New York ?***

####**Instructions**

+ Créez une procédure stockée nommée `cuspPickupZoneShiftStats` qui accepte `@Borough nvarchar(30)` comme paramètre d'entrée et limite les enregistrements avec la valeur correspondante `Borough`.
+ Calculez le 'Shift'en passant le hour de PickupDate à la fonction `dbo.GetShiftNumber()`. Utilisez la fonction `DATEPART` pour sélectionner uniquement la hourpartie du fichier PickupDate.
+ Regroupez par jour de la semaine `PickupDate`, quart de travail et zone.
+ Trier par jour de la semaine `PickupDate` (avec le lundi en premier), équipe et `TotalRideMin`.

```python
-- Create the stored procedure
CREATE PROCEDURE dbo.cuspPickupZoneShiftStats
	-- Specify @Borough parameter
	@Borough nvarchar(30)
AS
BEGIN
SELECT
	DATENAME(WEEKDAY, PickupDate) as 'Weekday',
    -- Calculate the shift number
	dbo.GetShiftNumber(DATEPART(hour, PickupDate)) as 'Shift',
	Zone.Zone as 'Zone',
	FORMAT(AVG(dbo.ConvertDollar(TotalAmount, .77)/dbo.ConvertMiletoKM(TripDistance)), 'c', 'de-de') AS 'AvgFarePerKM',
	FORMAT(COUNT (ID),'n', 'de-de') as 'RideCount',
	FORMAT(SUM(DATEDIFF(SECOND, PickupDate, DropOffDate))/60, 'n', 'de-de') as 'TotalRideMin'
FROM YellowTripData
INNER JOIN TaxiZoneLookup as Zone on PULocationID = Zone.LocationID
WHERE
	dbo.ConvertMiletoKM(TripDistance) > 0 AND
	Zone.Borough = @Borough
GROUP BY
	DATENAME(WEEKDAY, PickupDate),
    -- Group by shift
	dbo.GetShiftNumber(DATEPART(hour, PickupDate)),
	Zone.Zone
ORDER BY CASE WHEN DATENAME(WEEKDAY, PickupDate) = 'Monday' THEN 1
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Tuesday' THEN 2
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Wednesday' THEN 3
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Thursday' THEN 4
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Friday' THEN 5
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Saturday' THEN 6
              WHEN DATENAME(WEEKDAY, PickupDate) = 'Sunday' THEN 7 END,
         -- Order by shift
         dbo.GetShiftNumber(DATEPART(hour, PickupDate)),
         SUM(DATEDIFF(SECOND, PickupDate, DropOffDate))/60 DESC
END;
```
