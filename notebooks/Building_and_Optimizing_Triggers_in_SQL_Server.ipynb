####**2. Qu’est-ce qu’un déclencheur?**

+ **Un trigger est un type spécial de procédure stockée qui est automatiquement exécuté lorsque des événements (tels que des modifications de données) se produisent sur le serveur de base de données.**

####**3. Types de déclencheurs (basés sur les commandes T-SQL)**

+ En termes de commandes Transact-SQL (T-SQL), il existe trois types de déclencheurs dans SQL Server :

  - **Les déclencheurs `DML (Data Manipulation Language)` sont exécutés lorsqu’un utilisateur ou un processus modifie des données via une instruction `INSERT`, `UPDATE` ou `DELETE`**.
    + **Ces déclencheurs sont associés à des instructions liées à des tables ou des vues**.
  - **Le langage `DDL (Data Definition Language)` déclenche le déclenchement en réponse à des instructions exécutées au niveau de la base de données ou du serveur, telles que `CREATE`, `ALTER` ou `DROP`**.
  - **L’ouverture de session déclenche un déclenchement en réponse à des événements `LOGON` lorsque la session d’un utilisateur est établie**.
+ Nous couvrirons tous ces types de déclencheurs plus en détail dans les leçons ultérieures.

####**4. Types de déclencheurs (en fonction du comportement)**

+ Une autre façon de classer les déclencheurs est basée sur leur comportement.

+ Un déclencheur peut se comporter différemment par rapport à l’instruction qui le déclenche, ce qui entraîne deux types de déclencheurs.

  + **Le premier type, un déclencheur `AFTER`, est utilisé lorsque vous souhaitez exécuter un morceau de code après l’instruction initiale qui déclenche le déclencheur**.
    + Un exemple de cas d’utilisation de ce type de déclencheur consiste à reconstruire un index après un grand ajout de données dans une table.
    + Un autre exemple consiste à utiliser un déclencheur pour envoyer des alertes lorsque des instructions `UPDATE` sont exécutées sur la base de données. Vous verrez plus d’exemples au fur et à mesure que vous progresserez dans le cours.

####**5. Types de déclencheurs (en fonction du comportement)**
+ Le deuxième type de déclencheur est le déclencheur `INSTEAD OF`.

+ **Comme son nom l’indique, un déclencheur `INSTEAD OF` n’effectuera pas l’opération initiale, mais exécutera du code personnalisé à la place**.

  + Voici quelques exemples d’utilisation de ce comportement pour empêcher l’insertion de données dans des tables, **empêcher les mises à jour ou les suppressions, ou même empêcher la suppression de tables**.
  + Vous pouvez informer l’administrateur de base de données d’un comportement suspect tout en empêchant toute modification.
  + Comme pour les déclencheurs `AFTER`, il existe plus de cas d’utilisation possibles que les exemples fournis
  + ici. Voyons maintenant comment nous créons réellement un déclencheur.

####**6. Définition du déclencheur (avec AFTER)**

+ **Étant donné qu’un `déclencheur` est un objet SQL Server, nous en ajoutons un nouveau à l’aide de l’instruction `CREATE`**.

+ Dans ce cas, nous utilisons l’instruction `CREATE TRIGGER` pour ajouter un nouveau déclencheur `DML`.

+ Après avoir nommé le déclencheur, nous devons le lier à un objet existant; dans cet exemple, il s’agit de la table « Products ». La partie suivante se concentre sur la définition du comportement du déclencheur. Nous utilisons un déclencheur `AFTER` pour cet exemple.

+ **Nous devons également spécifier l’événement qui déclenche le déclencheur.**

+ **Il peut s’agir d’une instruction `INSERT`, `UPDATE` ou `DELETE`. Tenons-nous-en à un `INSERT` pour l’instant. Le mot-clé AS marque le début de notre flux de travail déclencheur. Les actions effectuées par le déclencheur sont mentionnées après ce mot-clé.**

####**7. Définition du déclencheur (avec INSTEAD OF)**
+ **Ici, un autre exemple de déclencheur DML utilise une instruction `INSTEAD OF` pour empêcher les suppressions de la table**.
+ **Toute tentative de suppression de lignes de la table échouera en raison de l’utilisation de `INSTEAD OF`.**

####**8. APRÈS VS AU LIEU DE**
+ **Comprendre la différence entre les comportements `AFTER` et `INSTEAD OF` est très important**.
+ **Lorsque le mot clé `AFTER` est utilisé, le déclencheur n’effectue les actions définies qu’une fois l’événement qui l’a déclenché terminé**.
+ **Lorsque `INSTEAD OF` est utilisé, l’événement qui a déclenché le déclencheur ne s’exécutera pas et seules les actions de déclenchement seront exécutées.**

```python
-- Set up a new trigger
CREATE TRIGGER OrdersUpdatedRows
ON ORDERS
-- The trigger should fire after UPDATE statements
AFTER UPDATE
-- Add the AS keyword before the trigger body
AS
	-- Insert details about the changes to a dedicated table
	INSERT INTO OrdersUpdate(OrderID, OrderDate, ModifyDate)
	SELECT OrderID, OrderDate, GETDATE()
	FROM inserted;
```

####**2. Déclencheurs et procédures stockées**

+ **La première comparaison des déclencheurs concerne les procédures stockées régulières**.

+ **Le déclencheur est également une procédure stockée, mais d’un type particulier**.

+ **Qu’est-ce qui le rend si spécial?**
  + **Un déclencheur est déclenché lorsqu’un événement de base de données ou de serveur défini se produit**.
  + **Les procédures stockées régulières s’exécutent lorsqu’elles sont appelées explicitement**.
  + **Un déclencheur s’exécute uniquement lorsque l’événement qui le déclenche démarre**.
  + Les triggers n’autorisent pas les paramètres d’entrée ou les instructions de transaction (comme `BEGIN TRANSACTION` ou `COMMIT`)**. Ils ne peuvent pas non plus renvoyer de valeurs de sortie.
  + D’autre part, les procédures stockées régulières permettent des paramètres d’entrée, des instructions de transaction et la possibilité de renvoyer une sortie.

####**3. Déclencheurs et procédures stockées**

+ Ces considérations ont un impact sur les cas d’utilisation réels des déclencheurs et des procédures stockées régulières.

+ **Nous utilisons des déclencheurs pour l’audit et pour renforcer l’intégrité dans la plupart des cas, et nous utilisons des procédures stockées régulières pour des tâches plus générales.**

####**4. Déclencheurs et colonnes calculées**

+ La deuxième comparaison des déclencheurs concerne les colonnes calculées.

+ Les colonnes calculées sont un bon moyen d’automatiser le calcul des valeurs contenues dans certaines colonnes.

+ Les valeurs de colonne calculées sont déterminées en fonction des valeurs d’autres colonnes, mais uniquement de la même table.

+ Cette limitation peut être surmontée en utilisant des déclencheurs.

+ Un déclencheur peut également utiliser des colonnes d’autres tables pour calculer des valeurs.

+ Bien que ce calcul soit effectué avec les instructions `INSERT` ou `UPDATE` lors de l’utilisation d’un trigger, pour une colonne calculée, il fait partie de la définition de la table.

####**2. Définition et propriétés**

+ ***La manière la plus simple d'illustrer les propriétés spécifiques des déclencheurs `DDL` consiste à les comparer aux déclencheurs `DML`. La plus grande différence est indiquée par les noms des types de déclencheurs. Les `_déclencheurs DML_` répondent aux événements associés aux instructions `DML : INSERT, UPDATE et DELETE`. D'autre part, les déclencheurs `DDL` répondent aux événements associés aux instructions `DDL telles que CREATE, ALTER et DROP`. Alors que les déclencheurs DML peuvent être utilisés avec les mots clés `AFTER et INSTEAD OF`, les déclencheurs `DDL` n'acceptent que le mot clé `AFTER`. De plus, nous pouvons utiliser les tables spéciales "insérées" et "supprimées" lorsque nous travaillons avec des déclencheurs `DML`, mais il n'y a pas de tables spéciales à utiliser avec les déclencheurs `DDL`.***

+ ***En effet, pour les déclencheurs `SQL Server`, les mots clés `FOR` et `AFTER` ont le même résultat. Pour minimiser le risque de confusion, les utilisateurs utilisent souvent le mot clé `FOR` pour les déclencheurs `DDL` et le mot clé `AFTER` pour les déclencheurs `DML`. Nous nous en tiendrons à cette convention ici, mais sachez que les deux versions sont correctes et se comporteront de la même manière.***

####**Audit de table à l'aide de déclencheurs**

+ ***La société a décidé de suivre les modifications apportées à toutes les tables les plus importantes. L'une de ces tables est Orders.***

```python
-- Add a trigger that tracks table changes
CREATE TRIGGER OrdersAudit
ON Orders
AFTER INSERT, UPDATE, DELETE
AS
	DECLARE @Insert BIT = 0;
	DECLARE @Delete BIT = 0;
	IF EXISTS (SELECT * FROM inserted) SET @Insert = 1;
	IF EXISTS (SELECT * FROM deleted) SET @Delete = 1;
	INSERT INTO OrdersAudit (TableName, EventType, UserAccount, EventDate)
	SELECT 'Orders' AS TableName
	       ,CASE WHEN @Insert = 1 AND @Delete = 0 THEN 'INSERT'
				 WHEN @Insert = 1 AND @Delete = 1 THEN 'UPDATE'
				 WHEN @Insert = 0 AND @Delete = 1 THEN 'DELETE'
				 END AS Event
		   ,ORIGINAL_LOGIN() AS UserAccount
		   ,GETDATE() AS EventDate;
```

####**Instructions**

+ Ajoutez un nouveau déclencheur qui se déclenche pour les déclarations `INSERT` et vérifie si la quantité de la commande peut être satisfaite par le stock actuel.
+ Génère une erreur si le stock est insuffisant. Sinon, effectuez un `INSERT` en utilisant la insertedtable spéciale.

```python
-- Create a new trigger to confirm stock before ordering
CREATE TRIGGER ConfirmStock
ON Orders
INSTEAD OF INSERT
AS
	IF EXISTS (SELECT *
			   FROM Products AS p
			   INNER JOIN inserted AS i ON i.Product = p.Product
			   WHERE p.Quantity < i.Quantity)
	BEGIN
		RAISERROR ('You cannot place orders when there is no stock for the order''s product.', 16, 1);
	END
	ELSE
	BEGIN
		INSERT INTO Orders (OrderID, Customer, Product, Price, Currency, Quantity, WithDiscount, Discount, OrderDate, TotalAmount, Dispatched)
		SELECT OrderID, Customer, Product, Price, Currency, Quantity, WithDiscount, Discount, OrderDate, TotalAmount, Dispatched FROM inserted;
	END;
```

```python
SELECT t.name AS TriggerName
FROM sys.objects AS o
INNER JOIN sys.triggers AS t ON t.parent_id = o.object_id
-- Get the trigger events
INNER JOIN sys.trigger_events AS te ON te.object_id = t.object_id
WHERE o.name = 'Orders'
-- Filter for triggers reacting to new rows
AND te.type_desc = 'UPDATE';
```
