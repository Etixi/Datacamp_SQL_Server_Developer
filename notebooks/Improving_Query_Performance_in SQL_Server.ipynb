####**2. Qu’est-ce que l’aliasing ?**
+ ***L’aliasing fait référence à la fourniture d’un nom court, d’un code ou d’une identité simple aux tables, colonnes et sous-requêtes d’une requête. Les alias sont temporaires et ne sont appliqués que pendant l’exécution de la requête ; Ils ne modifient pas les noms de table ou de colonne dans la base de données. Ils peuvent être pratiques pour rendre le code d’une requête plus facile à lire. Et, dans de nombreux cas, sont nécessaires, par exemple, une nouvelle colonne ou sous-requête, qui sera réutilisée dans une autre partie de la requête. Examinons quelques raisons pour lesquelles nous devons utiliser le aliasing.***

####**3. Pourquoi utiliser l'Aliasing ?**
+ Nous utilisons le repliement pour :
  + **Évitez l’utilisation répétitive de noms de table ou de colonne longs.**
  + **Identifiez facilement les tables jointes et les colonnes associées.**
  + **Identifiez les nouvelles colonnes.**
  + **Identifiez les sous-requêtes.**
  + **Évitez toute ambiguïté lorsque les colonnes des tables jointes partagent le même nom.**
  + **Et renommez les colonnes.**

####**Aliasing - équipe IMC**

+ ***Un statisticien du basketball aimerait connaître l'indice de masse corporelle (IMC) moyen par équipe de la NBA, en particulier toute équipe ayant un IMC moyen de 25 ou plus. Pour inclure Team dans la requête, vous devrez joindre la table Players à la table PlayerStats. La requête nécessitera un alias pour :***

  + Identifiez facilement les tables jointes et les colonnes associées.
  + Identifiez les sous-requêtes.
  + Éviter l'ambiguïté dans les noms de colonne.
  + Identifiez les nouvelles colonnes.


####**Instructions**

+ Attribuez à la nouvelle colonne IMC moyen le nom d'AvgTeamBMI.
+ Alias la table PlayerStats comme ps.
+ Alias la sous-requête comme p.
+ La table PlayerStats et la sous-requête se rejoignent sur la colonne PlayerName.
+ Ajoutez les alias aux colonnes PlayerName jointes.

```python
SELECT Team,
   ROUND(AVG(BMI),2) AS AvgTeamBMI -- Alias the new column
FROM PlayerStats AS ps -- Alias PlayerStats table
INNER JOIN
		(SELECT PlayerName, Country,
			Weight_kg/SQUARE(Height_cm/100) BMI
		 FROM Players) AS p-- Alias the sub-query
             -- Alias the joining columns
	ON ps.PlayerName = p.PlayerName
GROUP BY Team
HAVING AVG(BMI) >= 25;
```

+ **Techniquement oui, c’est une sous-requête corrélée**. + **Toutefois, EXISTS arrêtera la recherche dans la sous-requête lorsque la condition est TRUE alors que IN, avec une sous-requête, collectera tous les résultats de la sous-requête avant de poursuivre le traitement, ce qui est potentiellement plus lent**.

####**1. Indices**

####**2. Qu’est-ce qu’un indice?**
+ **Un index est une structure ajoutée à une table pour améliorer la vitesse d’accès aux données**.

+ **Les index sont utilisés pour localiser rapidement les données sans avoir à balayer ou à rechercher chaque ligne d’une table, ce qui les rend particulièrement utiles pour améliorer les performances des requêtes avec des conditions de filtre**.

+ **Les index sont appliqués aux colonnes de table et peuvent être ajoutés à tout moment**.

+ **L’ajout d’un index à une table est généralement effectué par l’administrateur de base de données**.

+ **SQL Server prend en charge plusieurs types d’index différents. Les deux index de table les plus courants sont Clustered et Nonclustered.**

####**3. Index clusterisés et non clusterisés**

+ **Une bonne analogie pour un index groupé est un dictionnaire où les mots sont stockés par ordre alphabétique. **

+ **Avec un index clusterisé, les données, dans les pages de données, sont stockées et classées par colonne ou par colonne avec l’index.**

+ **Étant donné que les données sont triées dans un sens, une table ne peut avoir qu’un seul index clusterisé**.

+ **Les index clusterisés réduisent le nombre de pages de données lues par une requête, ce qui permet d’accélérer les opérations de recherche.**

####**4. Index clusterisés et non clusterisés**


+ **Pour un index non clusterisé, une bonne analogie est un manuel avec un index à la fin**.

+ **Les données du livre ne sont pas ordonnées et l’index à l’arrière indique les numéros de page contenant une condition de recherche. **

+ **Comme un manuel, une table avec un index non clusterisé permet de désordonner les données dans les pages de données de table.**

+ **Une autre couche de la structure d’index contient des pointeurs ordonnés vers les pages de données.**

+ **Une table peut contenir plusieurs index non clusterisés.**

+ **Les index non clusterisés sont couramment utilisés pour améliorer les opérations d’insertion et de mise à jour de table.**

####**5. Index clusterisé : structure de l’arborescence B**

+ Voyons comment un index clusterisé peut améliorer les performances.

+ **Un index clusterisé crée ce que l’on appelle une arborescence B sur une table**.

+ Nous pouvons considérer un arbre B comme un arbre à l’envers où le tronc est appelé le nœud racine, les branches des nœuds de branche et quitte les nœuds de page.

####**6. Index clusterisé : structure de l’arborescence B**

+ Le nœud racine contient des pointeurs ordonnés vers des nœuds de branche qui, à leur tour, contiennent des pointeurs ordonnés vers des nœuds de page.

+ Le niveau du nœud de page contient toutes les pages de données de 8 kilo-octets de la table avec les données physiquement ordonnées par la ou les colonnes avec l’index clusterisé.

+ Sans index clusterisé, il n’y a aucune garantie que les données sont classées dans les pages de données de table.
