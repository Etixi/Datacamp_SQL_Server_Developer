####**2. La contrainte non nulle**

+ Comme son nom l’indique déjà, la contrainte `not-null` interdit toute valeur `« NULL »` sur une colonne donnée.

+  Cela doit être vrai pour l’état existant de la base de données, mais aussi pour tout état futur.

+ Par conséquent, vous ne pouvez spécifier une contrainte non nulle que sur une colonne qui ne contient pas encore de valeurs `« NULL »`.

+ Et: Il ne sera pas possible d’insérer des valeurs `« NULL »` à l’avenir.

####**3. Que signifie pour les valeurs NULL ?**

+ Avant de continuer à expliquer comment spécifier des contraintes non nulles, je veux que vous réfléchissiez aux valeurs « NULL ».

+ Qu’est-ce qu’ils signifient réellement pour vous?

+ Il n’y a pas de définition claire.

+ `« NULL »` peut signifier deux choses, par exemple, que la valeur est `inconnue ou n’existe pas du tout`. Il est également possible qu’une valeur ne s’applique pas à la colonne. Prenons un exemple.

####**4. Que signifie pour les valeurs `NULL` ? Un exemple**
+ Disons que nous définissons une table « étudiants ».

+ Les deux premières colonnes pour le numéro de sécurité sociale et le nom de famille ne peuvent pas être « NULL », ce qui est logique:
  + cela devrait être connu et s’appliquer à chaque étudiant.
+ Les colonnes « home_phone » et « office_phone » devraient cependant permettre des valeurs nulles - ce qui est la valeur par défaut, soit dit en passant.

+ Pourquoi? Tout d’abord, ces chiffres peuvent être inconnus, pour n’importe quelle raison, ou tout simplement ne pas exister, parce qu’un étudiant pourrait ne pas avoir de téléphone.

+ De plus, certaines valeurs ne s’appliquent tout simplement pas : certains étudiants peuvent ne pas avoir de bureau, donc ils n’ont pas de téléphone de bureau, et ainsi de suite.

+ Ainsi, une chose importante à retenir est que deux valeurs `« NULL »` ne doivent pas avoir la même signification.

+ Cela signifie également que la comparaison de `« NULL »` avec `« NULL »` donne toujours une valeur `« FALSE »`.

####**5. Comment ajouter ou supprimer une contrainte non nulle**
+ Vous venez de voir comment ajouter une contrainte non nulle à certaines colonnes lors de la création d’une table.

####**6. La contrainte unique**

+ ***La contrainte d’unicité sur une colonne garantit qu’il n’y a pas de doublons dans une colonne**.

+ Tout comme avec la contrainte `non-null`, vous ne pouvez ajouter une contrainte d’unicité que si la colonne ne contient pas de doublons avant de l’appliquer.

####**4. Qu’est-ce qu’une clé?**

+ Avant d’entrer dans le vif du sujet de ce qu’est réellement une clé primaire, examinons les clés en général.

+ **En général, une table de base de données possède un attribut, ou une combinaison de plusieurs attributs, dont les valeurs sont uniques dans l’ensemble de la table**.

+ **Ces attributs identifient un enregistrement de manière unique**.

+ **Normalement, une table, dans son ensemble, ne contient que des enregistrements uniques, ce qui signifie que la combinaison de tous les attributs est une clé en soi.**

+ **Toutefois, il ne s’agit pas d’une clé, mais d’une superclé, si les attributs de cette combinaison peuvent être supprimés et que les attributs identifient toujours les enregistrements de manière unique**.

+ **Si tous les attributs possibles ont été supprimés mais que les enregistrements sont toujours identifiables de manière unique par les attributs restants, nous parlons d’une superclé minimale. C’est la clé réelle.**

+ **Donc, une clé est toujours minimale. Prenons un exemple.**

###**Clés de substitution**
####**1. Clés de substitution**

+ ***Les clés de substitution sont une sorte de clé primaire artificielle. En d’autres termes, ils ne sont pas basés sur une colonne native dans vos données, mais sur une colonne qui existe juste pour avoir une clé primaire. Pourquoi auriez-vous besoin de cela?***

####**2. Clés de substitution**

+ **Il y a plusieurs raisons pour créer une clé de substitution artificielle**.

+ **Comme mentionné précédemment, une clé primaire est idéalement construite à partir du moins de colonnes possible**.

+ **Deuxièmement, la clé primaire d’un enregistrement ne devrait jamais changer au fil du temps**.

+ **Si vous définissez une clé primaire artificielle, idéalement constituée d’un nombre ou d’une chaîne unique, vous pouvez être sûr que ce nombre reste le même pour chaque enregistrement. D’autres attributs peuvent changer, mais la clé primaire a toujours la même valeur pour un enregistrement donné.**

###**3. Un exemple**

+ Il n’y a donc pas besoin d’une clé de substitution ici. Cependant, disons qu’il n’y avait que ces trois attributs dans le tableau.

+ La seule clé primaire sensée serait la combinaison de « make » et « model », mais c’est deux colonnes pour la clé primaire.

####**4. Ajout d’une clé de substitution avec le type de données série**
+ **Vous pouvez ajouter une nouvelle colonne clé de substitution, appelée « id », pour résoudre ce problème**.

+ **En fait, il existe un type de données spécial dans PostgreSQL qui permet d’ajouter des nombres auto-incrémentés à une table existante: le type « série »**.

+ Il est spécifié comme n’importe quel autre type de données.

+ Une fois que vous avez ajouté une colonne de type « série », tous les enregistrements de votre table seront numérotés.

+ Chaque fois que vous ajoutez un nouvel enregistrement à la table, il obtient automatiquement un nombre qui n’existe pas encore.

+ Il existe des types de données similaires dans d’autres systèmes de gestion de base de données, comme MySQL.

####**5. Ajout d’une clé de substitution avec le type de données série (suite)**

+ En outre, si vous essayez de spécifier un `ID` qui existe déjà, la contrainte de clé primaire vous empêchera de le faire.

+ Donc, après tout, la colonne « id » identifie de manière unique chaque enregistrement de cette table – ce qui est très utile, par exemple, lorsque vous souhaitez faire référence à ces enregistrements à partir d’une autre table. Mais ce sera le sujet du prochain chapitre.

####**6. Un autre type de clé de substitution**

+ Une autre stratégie pour créer une clé de substitution consiste à combiner deux colonnes existantes en une nouvelle.

+ Dans cet exemple, nous ajoutons d’abord une nouvelle colonne avec le type de données `« varchar »`.

+ Nous `« ACTUALISONS »` ensuite cette colonne avec la concaténation de deux colonnes existantes.

+ La fonction `« CONCAT »` colle ensemble les valeurs de deux ou plusieurs colonnes existantes. Enfin, nous transformons cette nouvelle colonne en clé primaire de substitution.

####**1. Intégrité référentielle**

+ ***Nous allons maintenant parler d'un des concepts les plus importants dans les systèmes de bases de données : l'intégrité référentielle. C'est un concept très simple...***

####**2. Intégrité référentielle**

+ ***... qui stipule qu'un enregistrement faisant référence à un autre enregistrement dans une autre table doit toujours faire référence à un enregistrement existant. En d'autres termes : un enregistrement de la table A ne peut pointer vers un enregistrement de la table B qui n'existe pas. L'intégrité référentielle est une contrainte qui concerne toujours deux tables, et est imposée par des clés étrangères, comme vous l'avez vu dans les leçons précédentes de ce chapitre. Ainsi, si vous définissez une clé étrangère dans la table "professors" référençant la table "universities", l'intégrité référentielle est maintenue de "professors" à "universities".***

####**3. Violations de l'intégrité référentielle**

+ ***L'intégrité référentielle peut être violée de deux manières. Supposons que la table A fasse référence à la table B. Ainsi, si un enregistrement de la table B qui est déjà référencé à partir de la table A est supprimé, vous avez une violation. D'autre part, si vous essayez d'insérer un enregistrement dans la table A qui fait référence à quelque chose qui n'existe pas dans la table B, vous violez également le principe. Et c'est la raison principale des clés étrangères - elles généreront des erreurs et vous empêcheront de faire ces choses accidentellement.***

####**4. Traiter les infractions**

+ ***Cependant, lancer une erreur n'est pas la seule option. Si vous spécifiez une clé étrangère sur une colonne, vous pouvez en fait indiquer au système de base de données ce qui doit se passer si une entrée de la table référencée est supprimée. Par défaut, le mot clé "ON DELETE NO ACTION" est automatiquement ajouté à une définition de clé étrangère, comme dans l'exemple ici. Cela signifie que si vous essayez de supprimer un enregistrement dans la table B qui est référencé à partir de la table A, le système générera une erreur. Cependant, il existe d'autres options. Par exemple, il y a l'option "CASCADE", qui permettra d'abord la suppression de l'enregistrement dans la table B, puis supprimera automatiquement tous les enregistrements de référence dans la table A. Ainsi, cette suppression est en cascade.***

####**5. Traitement des violations, suite.**

+ ***Il y a encore plus d'options. L'option "RESTRICT" est presque identique à l'option "NO ACTION". Les différences sont techniques et dépassent le cadre de ce cours. Plus intéressante est l'option "SET NULL". Il définira la valeur de la clé étrangère pour cet enregistrement sur "NULL". L'option "SET DEFAULT" ne fonctionne que si vous avez spécifié une valeur par défaut pour une colonne. Il modifie automatiquement la colonne de référence à une certaine valeur par défaut si l'enregistrement référencé est supprimé. La définition des valeurs par défaut dépasse également le cadre de ce cours, mais cette option est toujours bonne à connaître.***

####**Qu’est-ce que la normalisation de base de données ?**

+ ***La normalisation de base de données est un principe de conception de base de données permettant d’organiser les données de manière organisée et cohérente.***

+ **Il vous aide à éviter la redondance et à maintenir l’intégrité de la base de données. Il vous aide également à éliminer les caractéristiques indésirables associées à l’insertion, à la suppression et à la mise à jour.**

####**Quel est le but de la normalisation?**

+ ***L’objectif principal de la normalisation de base de données est d’éviter les complexités, d’éliminer les doublons et d’organiser les données de manière cohérente. Dans la normalisation, les données sont divisées en plusieurs tables liées entre elles par des relations.***

+ **Les administrateurs de base de données peuvent établir ces relations à l’aide de clés primaires, de clés étrangères et de clés composites.**

+ Pour ce faire, une clé primaire dans une table, par exemple, est liée à la valeur d’une autre table, par exemple, .employee_wagesemployee_data

+ **`N.B. : Une clé primaire est une colonne qui identifie de manière unique les lignes de données de cette table.`**

  + *Il s’agit d’un identifiant unique tel qu’un identifiant d’employé, un identifiant d’étudiant, un numéro d’identification de l’électeur (NIV), etc.*

+ **Une clé étrangère est un champ qui se rapporte à la clé primaire d’une autre table.**

Une clé composite est comme une clé primaire, mais au lieu d’avoir une colonne, elle a plusieurs colonnes.

####**Qu’est-ce que `1NF 2NF et 3NF` ?**
+ **`1NF`, `2NF` et `3NF` sont les trois premiers types de normalisation de base de données. Ils représentent respectivement la première forme normale, la deuxième forme normale et la troisième forme normale.**

+ **Il existe également `4NF` (quatrième forme normale) et `5NF` (cinquième forme normale). Il y a même 6NF (sixième forme normale), mais la forme normale la plus courante que vous verrez  est `3NF` (troisième forme normale).**

+ **Tous les types de normalisation de base de données sont cumulatifs, ce qui signifie que chacun s’appuie sur ceux qui se trouvent en dessous. Ainsi, tous les concepts de 1NF sont également transférés à 2NF, et ainsi de suite.**

####**La première forme normale – 1NF**
+ **Pour qu’une table soit dans la première forme normale, elle doit répondre aux critères suivants :**

  + **Une seule cellule ne doit pas contenir plus d’une valeur (atomicité)**
  + il doit y avoir une clé primaire pour l’identification
  + Pas de lignes ou de colonnes dupliquées
  + Chaque colonne ne doit avoir qu’une seule valeur pour chaque ligne de la table

####**La deuxième forme normale – 2NF**
+ **Le 1NF élimine uniquement les groupes répétitifs, pas la redondance. C’est pourquoi il y a 2NF.**

+ **Une table est dite en 2NF si elle répond aux critères suivants :**

  + c’est déjà en 1NF
  + n’a pas de dépendance partielle. Autrement dit, tous les attributs non clés dépendent entièrement d’une clé primaire.

####**La troisième forme normale – 3NF**
+ **Lorsqu’une table est en 2NF, elle élimine les groupes répétitifs et la redondance, mais elle n’élimine pas la dépendance partielle transitive.**

+ **Cela signifie qu’un attribut non premier (un attribut qui ne fait pas partie de la clé du candidat) dépend d’un autre attribut non premier. C’est ce que la troisième forme normale (3NF) élimine.**


+ Ainsi, pour qu’une table soit en 3NF, elle doit :

  + être en 2NF
  + n’ont pas de dépendance partielle transitive.
